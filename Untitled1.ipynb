```python
# 'Cabin' has excessive missing data (~77%); insufficient for imputation or modeling — dropping it
data.drop(columns=['Cabin'], inplace=True)
```

```python
data.isnull().sum()
```

```python
# Imputing the single missing value in 'Fare' with the median fare
data['Fare'].fillna(data['Fare'].median(), inplace=True)
```

```python
# Extracting honorific titles (e.g., Mr, Miss, Mrs) from 'Name' to create a new 'Title' feature.
# Then, imputing missing 'Age' values using the median age of passengers sharing the same title — 
# this provides more context-aware and accurate imputation than a global median.
data['Title'] = data['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
data['Age'] = data.groupby('Title')['Age'].transform(lambda x: x.fillna(x.median()))
```

```python
numeric_cols = ['Age', 'Fare', 'SibSp', 'Parch']
plt.figure(figsize=(12, 8))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(data[col], color='skyblue')
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# List of columns to plot
num_cols = ['Age', 'Fare', 'SibSp', 'Parch']

# Set up a 2x2 grid
plt.figure(figsize=(12, 8))

for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 2, i)
    
    if col in data.columns and data[col].dropna().shape[0] > 0:
        sns.histplot(data[col], bins=30, kde=True, color='steelblue')
        plt.title(f'Distribution of {col}')
    else:
        plt.text(0.5, 0.5, f'{col} data missing', ha='center', va='center')
        plt.xticks([])
        plt.yticks([])

plt.tight_layout()
plt.show()
```
